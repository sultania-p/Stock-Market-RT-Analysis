```python
import pandas as pd
from time import sleep
from kafka import KafkaConsumer
from json import dumps
import json
from s3fs import S3FileSystem
```

```python
consumer = KafkaConsumer('demo_test',
                         bootstrap_servers=['54.227.205.231:9092'],
                         value_deserializer=lambda x: json.loads(x.decode('utf-8')))
```

```python
for c in consumer:
    print(c.value)
```

```python
s3 = S3FileSystem()
```

```python
for count, i in enumerate(consumer):
    with s3.open("s3://kafka-stock-market-real-time-analysis-sultania.p/stock_market_{}.json".format(count), 'w') as file:
        json.dump(i.value, file)
```
